```python
%matplotlib inline
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1)
x = np.linspace(0,
                150, 1000)
ax.plot(x, norm.pdf(x, loc=50, scale=10),
       'r-', lw=5, alpha=0.6, label='prior')
ax.plot(x, norm.pdf(x, loc=40, scale=100),
       'b-', lw=5, alpha=0.6, label='likelihood')
ax.plot(x, norm.pdf(x, loc=49.09, scale=100/11),
       'g-', lw=5, alpha=0.6, label='likelihood')
```
